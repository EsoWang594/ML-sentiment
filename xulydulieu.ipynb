1.1 Lam sach co ban

```python
def clean_text(text):
  text = str(text).lower() #viet thuong tat ca chu
  text = re.sub(r"http\S+|www|s+|https\S+" , '', text , flags=re.MULTILINE) #xoas URL
  text = re.sub(r"|@w+|\#" , '',text )# xoa @mention , hashtag
  text = re.sub(r"[^\w\s]" , '', text) #xoa dau cau
  text = re.sub(r'[a-zA-Zà-ỹÀ-ỸX0-9\s]' , '',text)#giu lai chu va so
  text = re.sub(r'\s+', '' , text).strip() #chuan hoa khoang trang
  return text

df ['clean_data'] = df['Comment'].apply(clean_text)
df.head(30)
```

1.2 Loại bỏ spam

```python
def is_spam(text):
    text = str(text).lower()
    if len(text) < 5:
        return True
    if re.search(r"(https?://|www\.|@|#|\.com)", text):
        return True
    if re.search(r"(.)\1{4,}", text):
        return True
    if sum(c.isdigit() for c in text) > len(text) * 0.5:
        return True
    if len(re.findall(r"[^\w\s]", text)) > len(text) * 0.4:
        return True
    return False

df["is_spam"] = df["Comment"].apply(is_spam)
df = df[df["is_spam"] == False]
```

1.3 Sửa lỗi chính tả

```python
typo_dict = {
    "ko": "không", "k": "không", "kg": "không", "hok": "không",
    "dc": "được", "đc": "được", "cx": "cũng", "mik": "mình",
    "bt": "bình thường", "vs": "với", "j": "gì", "qá": "quá", "wá": "quá",
}

def clean_comment(text):
    if not isinstance(text, str):
        return ""
    text = text.lower()
    for wrong, right in typo_dict.items():
        text = re.sub(rf'\b{wrong}\b', right, text)
    text = re.sub(r'(.)\1{2,}', r'\1', text)
    return text

df['cleaned_comment'] = df['Comment'].apply(clean_comment)
df.head(30)
```

```python
import pandas as pd
import re
from underthesea import word_tokenize

def load_stopwords(file_path):
    """
    Hàm đọc danh sách stopwords từ file.
    """
    with open("/content/vietnamese-stopwords.txt", 'r', encoding='utf-8') as f:
        stopwords = f.read().splitlines()  # Đọc từng dòng và lưu thành danh sách
    return stopwords

def preprocess_text(text, stopwords):
    """
    Hàm làm sạch và tối ưu hóa văn bản theo yêu cầu PhoBERT.
    """
    # Loại bỏ ký tự đặc biệt
    text = re.sub(r'[^\w\s]', '', text)

    # Tách từ (word segmentation)
    tokenized_text = word_tokenize(text)

    # Loại bỏ stopwords
    filtered_text = [word for word in tokenized_text if word.lower() not in stopwords]

    # Kết quả word-segmented cho PhoBERT
    return ' '.join(filtered_text)

def optimize_csv(file_path, stopwords_path, output_path):
    """
    Hàm tối ưu hóa dữ liệu từ file CSV.
    """
    try:
        # Đọc danh sách stopwords
        stopwords = load_stopwords(stopwords_path)

        # Đọc file CSV
        df = pd.read_csv("/content/66k_rating0.csv")

        # Kiểm tra cột 'comment' có tồn tại không
        if 'Comment' not in df.columns:
            raise ValueError("File CSV không có cột 'comment'!")

        # Tiền xử lý dữ liệu trong cột 'comment'
        df['processed_comment'] = df['Comment'].apply(lambda x: preprocess_text(x, stopwords))

        # Lưu file CSV tối ưu hóa
        df.to_csv(output_path, index=False, encoding='utf-8')
        print(f"Dữ liệu đã được tối ưu hóa và lưu tại: {output_path}")

    except Exception as e:
        print(f"Đã xảy ra lỗi: {e}")

# File đầu vào
input_csv = '66k_rating0.csv'  # File CSV của bạn
stopwords_file = 'stopwords.txt'  # File chứa danh sách stopwords
output_csv = 'optimized_66k_rating.csv'

# Gọi hàm tối ưu hóa
optimize_csv(input_csv, stopwords_file, output_csv)
```

```python
from underthesea import word_tokenize
import torch

def preprocess_text_for_phobert(text):
    # Tách từ với underthesea
    tokenized_text = word_tokenize(text)
    # Chuyển thành dạng phù hợp với PhoBERT
    return ' '.join(tokenized_text)

def prepare_inputs(comments, tokenizer):
    # Mã hóa văn bản
    inputs = tokenizer(
        comments,
        padding=True,
        truncation=True,
        return_tensors="pt",
        max_length=128  # Giới hạn độ dài văn bản
    )
    return inputs
```

```python
def predict_sentiment(comments, model, tokenizer):
    # Tiền xử lý văn bản
    processed_comments = [preprocess_text_for_phobert(comment) for comment in comments]

    # Mã hóa
    inputs = prepare_inputs(processed_comments, tokenizer)

    # Dự đoán
    model.eval()
    with torch.no_grad():
        outputs = model(**inputs)
        predictions = torch.argmax(outputs.logits, dim=1)

    # Chuyển nhãn số sang nhãn cảm xúc
    sentiment_labels = {0: "Tiêu cực", 1: "Trung lập", 2: "Tích cực"}
    return [sentiment_labels[pred.item()] for pred in predictions]
```

```python
import re

def split_sentences(comment):
    # Tách dựa trên dấu phẩy, dấu chấm hoặc các từ liên kết phổ biến
    sentences = re.split(r',|\.|\snhưng\s|\svà\s', comment)
    return [sentence.strip() for sentence in sentences if sentence.strip()]

```

```python
# Hàm gán khía cạnh dựa trên từ điển từ khóa
def assign_aspect(sentence, keywords):
    for aspect, words in keywords.items():
        if any(word in sentence for word in words):
            return aspect
    return "khác"
```

```python
# Hàm dự đoán cảm xúc từ mô hình
def predict_sentiment(sentence, model, vectorizer):
    X_input = vectorizer.transform([sentence])  # Chuyển câu thành vector
    prediction = model.predict(X_input)
    sentiment_labels = {0: "tiêu cực", 1: "trung lập", 2: "tích cực"}
    return sentiment_labels[prediction[0]]

# Hàm tạo cột "khía cạnh và cảm xúc"
def create_aspect_sentiment_column(comment, model, vectorizer, keywords):
    sentences = split_sentences(comment)  # Tách câu
    unique_labels = {}
    for sentence in sentences:
        aspect = assign_aspect(sentence, keywords)
        if aspect not in unique_labels:  # Đảm bảo chỉ xuất hiện 1 lần
            sentiment = predict_sentiment(sentence, model, vectorizer)
            unique_labels[aspect] = f"{aspect}: {sentiment}"
    return ', '.join(unique_labels.values())  # Kết hợp các khía cạnh và cảm xúc
```

```python
# Tạo cột "khía cạnh và cảm xúc"
df["sentiment_aspect"] = df["cleaned_comment"].apply(
    lambda x: create_aspect_sentiment_column(x, loaded_model, vectorizer, keywords)
)
# Kiểm tra dữ liệu
print(df.head())
```
